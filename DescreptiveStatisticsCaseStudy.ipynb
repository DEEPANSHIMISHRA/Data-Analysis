```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Employee Dataset.csv')
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data = data,x = 'age', y = 'salary', color = 'red',edgecolor = 'linen', alpha = 0.5)
plt.title('Scatterplot of Employee Age vs Salary')
plt.xlabel('Employee Age')
plt.ylabel('Employee Salary')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data = data,x = 'healthy_eating', y = 'active_lifestyle', color = 'blue',edgecolor = 'linen', alpha = 0.5)
plt.title('Scatterplot of Healthy Eating vs Active Lifestyle')
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data = data,x = 'healthy_eating', y = 'salary', color = 'blue',edgecolor = 'k', alpha = 0.5)
plt.title('Scatterplot of Healthy Eating vs Employee Salary')
plt.xlabel('Healthy Eating')
plt.ylabel('Employee Salary')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data = data,x = 'groups',edgecolor = 'linen', alpha = 0.7)
plt.title('Count Plot of Groups')
plt.xlabel('Groups')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data = data,x = 'salary', color = 'orange',edgecolor = 'linen', alpha = 0.7,bins = 10)
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```
